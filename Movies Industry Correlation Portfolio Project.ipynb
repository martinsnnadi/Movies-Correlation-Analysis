```python
# import Libraries
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) #Adjust the configuration of the plot we create


# Read in the data

df = pd.read_csv(r'C:\Users\Nnadi Ifeanyichukwu\Desktop\BI\Portfolio\movies Industry\movies.csv')
```

```python
# let's see if there is any missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
df['budget'] = pd.to_numeric(df['budget'].astype(str).str.replace(',', ''), errors='coerce').fillna(0).astype('int64') 
df['gross'] = pd.to_numeric(df['gross'].astype(str).str.replace(',', ''), errors='coerce').fillna(0).astype('int64') 
```

```python
# Create New column and copy year into it from release
df['New_Year'] = df['released'].astype(str).str.split().str[2]
```

```python
# Sort data according to a defined column
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Scatter plot with budget vs gross
plt.scatter(x=df['budget'], y=df['gross'])

plt.title('budget vs Gross Earnings')

plt.xlabel('Gross Earnings')

plt.ylabel('Budget for film')

plt.show()
```

```python
# Plot the budget vs gross using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "red"}, line_kws={"color": "blue"})
```

```python
plt.show()
```

```python
# let's start looking at correlation, added (numeric_only = 'True') to the original df.corr to clear error

df.corr(numeric_only = 'True', method='kendall') #pearson, Kendall, spearman
```

```python
correlation_matrix = df.corr(numeric_only = 'True', method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
# This is how we gave all values in the df table a numeric value, so we will be able to run correlation on them

df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized
```

```python
correlation_matrix = df_numerized.corr(numeric_only = 'True', method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
df_numerized.corr()
```

```python
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
